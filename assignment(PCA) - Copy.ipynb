# PCA

```python
#importing libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
#importing dataset
df=pd.read_csv('wine.csv')
df
```

```python
df['Type'].value_counts()
```

```python
df1=df.iloc[:,1:]
df1
```

```python
df1.info()
```

```python
#Converting data to numpy array
df_arr=df1.values
df_arr
```

```python
# Normalizing the numerical data 
df_norm=scale(df_arr)
df_norm
```

# PCA implementation

```python
# Applying PCA Fit Transform to dataset
pca=PCA(n_components=13)
df_pca=pca.fit_transform(df_norm)
df_pca
```

```python
# PCA Components matrix or covariance Matrix
pca.components_
```

```python
# The amount of variance that each PCA has
df_var=pca.explained_variance_ratio_
df_var
```

```python
# Cummulative variance of each PCA
df_var1=np.cumsum(np.round(df_var,4)*100)
df_var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(df_var1, color='red')
```

```python
# Final Dataframe
final_df=pd.concat([df['Type'],pd.DataFrame(df_pca[:,0:3],columns=['PCA1','PCA2','PCA3'])],axis=1)
final_df
```

```python
# Visualization of PCAs
plt.figure(figsize=(12,10))
sns.scatterplot(data=final_df)
```

# Checking with other cluster algorithms

```python
#Hierarchical clustering
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(12,10))
dandogram=sch.dendrogram(sch.linkage(df_norm,'complete'))
```

```python
# Creating Clusters (y)
hcluster=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hcluster
```

```python
y=pd.DataFrame(hcluster.fit_predict(df_norm),columns=['clusterid'])
y['clusterid'].value_counts()
```

```python
# Adding clusters to dataset
df3=df1.copy()
df3['clusterid']=hcluster.labels_
df3
```

```python
#Kmean clustering
from sklearn.cluster import KMeans
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# Building cluster algorithm using k=3

```python
# Cluster algorithm using K=3
cluster3=KMeans(3,random_state=30).fit(df_norm)
cluster3
```

```python
cluster3.labels_
```

```python
# Assigning clusters to the data set
df4=df1.copy()
df4['cluster3id']=cluster3.labels_
df4
```
